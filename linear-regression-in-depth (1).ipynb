In the most simple words, Linear Regression is the supervised Machine Learning model in which the model finds the best fit linear line between the independent and dependent variable i.e it finds the linear relationship between the dependent and independent variable.
<br>
Linear Regression is of two types: Simple and Multiple.
    <br>
Simple Linear Regression is where only one independent variable is present and the model has to find the linear relationship of it with the dependent variable.
                  
<br>
Whereas, In Multiple Linear Regression there are more than one independent variables for the model to find the relationship.</p>

<a style="color:red" href="https://www.kaggle.com/mysarahmadbhat/types-of-transformations-for-better-distribution">TYPES OF TRANSFORMATIONS FOR BETTER DISTRIBUTION</a></p>

<p style="font-family: bold;font-size: 18px; text-align: center; color: Black">Removing some of the highly correlated independent variables.!</p>

<p style="font-family: Arials; font-size: 28px; text-align:center;font-style: bold; font-weight: bold; letter-spacing: 2px; color: #adc3ff; line-height:1.0">02. ENCODING.</p>

<p style="font-family: bold;font-size: 18px; text-align: center; color: Black">As we know our machine learning model takes only numerical values. But we have some of the categorical columns also let's change them!</p>

```python
bmw["transmission"]=bmw["transmission"].map({'Semi-Auto':0,'Automatic':1,'Manual':2})
bmw["fuelType"]=bmw["fuelType"].map({'Diesel':0,'Petrol':1,'Hybrid':2,'Other':3,'Electric':4})


```

```python
dum=pd.get_dummies(bmw["model"],drop_first=True)
dum2=pd.DataFrame(dum,columns=dum.columns)
df=pd.concat([bmw,dum2],axis=1)
df.drop("model",axis=1,inplace=True)
```

<p style="font-family: Arials; font-size: 28px; text-align:center;font-style: bold; font-weight: bold; letter-spacing: 2px; color: #adc3ff; line-height:1.0">CORRECT THE NORMALITY.</p>

<p style="font-family: bold;font-size: 18px; text-align: center; color: Black">We know that in the regression analysis the response variable should be normally distributed to get better prediction results.Also it is one of the assumption that we have to follow.From the Eda we get to know that there are so many outliers present in the daata set . Let's handle those also!
<br>
We will use Log Transformation</p>

```python
df["price"]=round(np.log(df["price"]))
df["year"]=round(np.log(df["year"]))
df["mileage"]=round(np.log(df["mileage"]))
df["tax"]=round(np.log(df["tax"]))
df["mpg"]=round(np.log(df["mpg"]))
df["engineSize"]=round(np.log(df["engineSize"]))
```

```python
X=df.drop("price",axis=1)
y=df["price"]
```

```python
df.head()
```

<p style="font-family: Arials; font-size: 28px; text-align:center;font-style: bold; font-weight: bold; letter-spacing: 2px; color: #adc3ff; line-height:1.0">Let's make the Model.</p>

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=0)


from sklearn.linear_model import LinearRegression
model=LinearRegression()
model.fit(X_train,y_train)

y_pred=model.predict(X_test)
```

<p style="font-family: bold;font-size: 18px; text-align: center; color: Black">Now we have made our linear regression model. But we neend to Evaluate the model! <br> To understand the performance of the Regression model performing model evaluation is necessary. </p>



<p id="part5"></p>
<p style="font-family: Arials; font-size: 28px; text-align:center;font-style: bold; font-weight: bold; letter-spacing: 2px; color: grey; line-height:1.0">05. Evaluation Metrics for Regression.</p>


<p style="font-family: Arials; font-size: 24px; font-style: bold; font-weight: bold; letter-spacing: 2px; color: #DC143C; line-height:1.0">01.  MEAN SQUARED ERROR: .</p>

<p style="font-family: bold;font-size: 18px; text-align: center; color: Black">The mean squared error tells you how close a regression line is to a set of points. It does this by taking the distances from the points to the regression line (these distances are the â€œerrorsâ€) and squaring them.</p>



<img src="https://cdn-images-1.medium.com/max/1600/1*3VJyfU1qBqoHwaDJm3KAKA.gif">

<p style="font-family: bold;font-size: 18px; text-align: center; color: Black">Where yi is the actual point and yi^ is the predicted.</p>

```python
from sklearn.metrics import mean_squared_error
mean_squared_error=mean_squared_error(y_test,y_pred)
mean_squared_error
```

<p style="font-family: Arials; font-size: 24px; font-style: bold; font-weight: bold; letter-spacing: 2px; color: #DC143C; line-height:1.0">02.  MEAN ABSOLUTE ERROR: .</p>

<img src="https://cdn-images-1.medium.com/max/640/1*iLabSjpdwd1TaZyKdDKYBA.png" width=500>

<p style="font-family: bold;font-size: 18px; text-align: center; color: Black">MAE is the average of the absolute difference between the predicted values and observed value!</p>

```python
from sklearn.metrics import mean_absolute_error
mean_absolute_error=mean_absolute_error(y_test,y_pred)
mean_absolute_error
```

<p style="font-family: Arials; font-size: 24px; font-style: bold; font-weight: bold; letter-spacing: 2px; color: #DC143C; line-height:1.0">03.  ROOT MEAN SQUARE ERROR: .</p>

<p style="font-family: bold;font-size: 18px; text-align: center; color: Black">It is the root of MSE i.e Root of the mean difference of Actual and Predicted values. RMSE has the benefit of penalizing large errors more so can be more appropriate in some cases!</p>

<img src="https://editor.analyticsvidhya.com/uploads/69457rmse.png" width=500>

```python
root_mean_squared_error=np.sqrt(mean_squared_error)
root_mean_squared_error
```

<p style="font-family: Arials; font-size: 24px; font-style: bold; font-weight: bold; letter-spacing: 2px; color: #DC143C; line-height:1.0">04.  R-SQUARE .</p>

<p style="font-family: bold;font-size: 18px; text-align: center; color: Black">The most commonly used metric for model evaluation in regression analysis is R squared. It can be defined as a Ratio of variation to the Total Variation. The value of R squared lies between 0 to 1, the value closer to 1 the better the model..</p>

<p style="font-family: Arials; font-size: 24px; font-style: bold; font-weight: bold; letter-spacing: 2px; color: #DC143C; line-height:1.0">05.  ADJUSTED R-SQUARE .</p>

<p style="font-family: bold;font-size: 18px; text-align: center; color: Black">It is the improvement to R squared. The problem/drawback with R2 is that as the features increase, the value of R2 also increases which gives the illusion of a good model. So the Adjusted R2 solves the drawback of R2. It only considers the features which are important for the model and shows the real improvement of the model.<br>
Adjusted R2 is always lower than R2!</p>

<img src="https://editor.analyticsvidhya.com/uploads/80741adjusted%20r2.png" width=400>
